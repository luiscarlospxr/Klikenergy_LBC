```python
#Path where the LBCs are at the moment
LBC_dir = New_directory
files = glob.iglob(os.path.join(LBC_dir, '*.xlsx'))


for file in files:
    print(file)

    # File directory
    FileDirectory = file
    
    # Base_line: read
    BaseLine = pd.read_excel(FileDirectory,index_col=0).reset_index()

    #Renombrar columnas de archivo virgen para ser proecesado en horas
    BaseLine.rename(columns={"Unidad Energía":"FECHA",'KWh':'H0','Unnamed: 2':'H1','Unnamed: 3':'H2','Unnamed: 4':'H3','Unnamed: 5':'H4','Unnamed: 6':'H5','Unnamed: 7':'H6','Unnamed: 8':'H7','Unnamed: 9':'H8','Unnamed: 10':'H9','Unnamed: 11':'H10','Unnamed: 12':'H11','Unnamed: 13':'H12','Unnamed: 14':'H13','Unnamed: 15':'H14','Unnamed: 16':'H15','Unnamed: 17':'H16','Unnamed: 18':'H17','Unnamed: 19':'H18','Unnamed: 20':'H19','Unnamed: 21':'H20','Unnamed: 22':'H21','Unnamed: 23':'H22','Unnamed: 24':'H23','Unnamed: 25':'TOTAL'}, inplace=True)

    #Base line: Columns selection  
    BaseLine = BaseLine[['FECHA','H0','H1','H2','H3','H4','H5','H6','H7','H8','H9','H10','H11','H12','H13','H14','H15','H16','H17','H18','H19','H20','H21','H22','H23']]
    #Base line: Last 105 days of measurements
    BaseLine = BaseLine.iloc[len(BaseLine['FECHA'])-61:len(BaseLine['FECHA'])-1].reset_index()
    #Reemplazar los valores NaN porceros.
    BaseLine = BaseLine.fillna(0)

    #Load Holidays
    # File directory
    FileDirectory = '../Colombian_Holidays/Holidays.xlsx'

    # Holidays: read
    HolidayList = pd.read_excel(FileDirectory,index_col=0).reset_index()
    
    #________________________________________________
    #1. Preparación de datos

    #1.1 Sumar consumo diario (Ct)
    #Base line: Daily consuption
    BaseLine['TOTAL'] = round(BaseLine['H0'] + BaseLine['H1'] + BaseLine['H2'] + BaseLine['H3'] + BaseLine['H4'] + BaseLine['H5'] + BaseLine['H6'] + BaseLine['H7'] + BaseLine['H8'] + BaseLine['H9'] + BaseLine['H10'] + BaseLine['H11'] + BaseLine['H12'] + BaseLine['H13'] + BaseLine['H14'] + BaseLine['H15'] + BaseLine['H16'] + BaseLine['H17'] + BaseLine['H18'] + BaseLine['H19'] + BaseLine['H20'] + BaseLine['H21'] + BaseLine['H22'] + BaseLine['H23'],2) 

    #1.2 Identificar qué día de la semana es cada fecha y procesa los festivos
    # Base line: Days convertion/correction
    BaseLine['WEEKDAY'] = pd.Series([0 for x in range(len(BaseLine['FECHA']))])
    BaseLine['FESTIVOS'] = pd.Series([0 for x in range(len(BaseLine['FECHA']))])

    #Asigna a cada fecha el día de la semana correspondiente
    for MeasuredDays in range(len(BaseLine['FECHA'])):
        BaseLine['WEEKDAY'].iloc[MeasuredDays] = datetime.strptime(BaseLine['FECHA'].iloc[MeasuredDays], '%Y-%m-%d').weekday() + 1


    #Identificar si el día es festivo (no domingo)
    for MeasuredDays in range(len(BaseLine['FECHA'])):
        if BaseLine['FECHA'][MeasuredDays] in list(HolidayList['Holidays']):
            BaseLine['FESTIVOS'][MeasuredDays] = 1
            BaseLine['WEEKDAY'][MeasuredDays] = 8

    #Reducción de datos para mayor velocidad
    BaseLine_summary    = BaseLine[['FECHA','WEEKDAY','TOTAL']]

    #2.1 Corrección de datos
    #DEPURACIÓN INICIAL

    #2.1.1 Dividir el data set por día de la semana
    BaseLine_LaboralDay = BaseLine_summary.loc[(BaseLine_summary['WEEKDAY'] == 1) |
                                               (BaseLine_summary['WEEKDAY'] == 2) |
                                               (BaseLine_summary['WEEKDAY'] == 3) |
                                               (BaseLine_summary['WEEKDAY'] == 4) |
                                               (BaseLine_summary['WEEKDAY'] == 5)]
    BaseLine_Saturdays   = BaseLine_summary.loc[BaseLine_summary['WEEKDAY'] == 6]
    BaseLine_Sundays     = BaseLine_summary.loc[BaseLine_summary['WEEKDAY'] == 7]
    BaseLine_Holydays    = BaseLine_summary.loc[BaseLine_summary['WEEKDAY'] == 8]

    #2.1.2 Contar candidad de días del mismo tipo
    Days_Laboral   = len(BaseLine_LaboralDay)
    Days_Saturday  = len(BaseLine_Saturdays)
    Days_Sunday    = len(BaseLine_Sundays)
    Days_Holyday   = len(BaseLine_Holydays)

    #2.1.3 Contar porcentaje de días con consumo 0 MWh
    Zero_LaboraDays = len(BaseLine_LaboralDay.loc[BaseLine_LaboralDay['TOTAL'] == 0])/Days_Laboral
    Zero_Saturdays  = len(BaseLine_Saturdays.loc[BaseLine_Saturdays['TOTAL'] == 0])/Days_Saturday
    Zero_Sundays    = len(BaseLine_Sundays.loc[BaseLine_Sundays['TOTAL'] == 0])/Days_Sunday
    Zero_Holydays   = len(BaseLine_Holydays.loc[BaseLine_Holydays['TOTAL'] == 0])/Days_Holyday

    #2.1.4 Para días tipo con más de 20% de registros con cero la LBC es 0
    if Zero_LaboraDays > 0.2:
        BaseLine_LaboralDay['TOTAL'] = 0
    if Zero_Saturdays > 0.2:
        BaseLine_Saturdays['TOTAL'] = 0
    if Zero_Sundays > 0.2:
        BaseLine_Sundays['TOTAL'] = 0
    if Zero_Holydays > 0.2:
        BaseLine_Holydays['TOTAL'] = 0

    #2.1.5 Completar día festivo (no domingo) con menos de 5 registros, por los valores menores del domingo.
    if Zero_Holydays < 5:
        ValueMin1 = BaseLine_Sundays['TOTAL'].nsmallest(1).iloc[-1] #El menor valor del domingo
        ValueMin2 = BaseLine_Sundays['TOTAL'].nsmallest(2).iloc[-1] #El segundo menor valor del domingo
        DateMin1  = BaseLine_Sundays.loc[BaseLine_Sundays['TOTAL'] == BaseLine_Sundays['TOTAL'].nsmallest(1).iloc[-1]]['FECHA'].to_numpy()[0] #Fecha en que ocurre el menor valor del domingo
        DateMin2  = BaseLine_Sundays.loc[BaseLine_Sundays['TOTAL'] == BaseLine_Sundays['TOTAL'].nsmallest(2).iloc[-1]]['FECHA'].to_numpy()[0] #Fecha en que ocurre el segundo menor valor del domingo
        min1 = [{'FECHA': DateMin1,'WEEKDAY':8,'TOTAL':ValueMin1}]
        min2 = [{'FECHA': DateMin2,'WEEKDAY':8,'TOTAL':ValueMin2}]
        BaseLine_Holydays = BaseLine_Holydays.append(min1,ignore_index=True,sort=False)
        BaseLine_Holydays = BaseLine_Holydays.append(min2,ignore_index=True,sort=False)

    #################################################
    #TRANSFORMACIÓN DE VALORES IGUALES A CERO
    #################################################

    #Días Laborales
    a = BaseLine_LaboralDay['TOTAL'].to_numpy()
    lenA =len(a)-1
    RangeA =  range(0,lenA)

    for n in RangeA:
        if a[n] == 0 and n == 0:
            BaseLine_LaboralDay['TOTAL'][n] = (a[n+1]+a[n+2])/2
        elif a[n] == 0 and a[n+1] == 0 and n == 1:
            BaseLine_LaboralDay['TOTAL'][n] = (a[n-1]+a[n+2])/2
        elif a[n] == 0:
            no_ceros = np.count_nonzero(pd.DataFrame({'TOTAL': [a[n-2],a[n-1],a[n+1],a[n+2]]})['TOTAL'].to_numpy())
            suma = sum(pd.DataFrame({'TOTAL': [a[n-2],a[n-1],a[n+1],a[n+2]]})['TOTAL'].to_numpy())
            media_movil = suma/(no_ceros)
            BaseLine_LaboralDay['TOTAL'][n] = media_movil      
        elif a[lenA] == 0 and n == lenA:
            BaseLine_LaboralDay['TOTAL'][n] = (a[n-1]+a[n-2])/2

    #TRANSFORMACIÓN DE VALORES IGUALES A CERO
    #Días Sábados
    f = BaseLine_Saturdays['TOTAL'].to_numpy()
    lenF =len(f)-1
    RangeF =  range(0,lenF)

    for n in RangeF:
        if f[n] == 0 and n == 0:
            BaseLine_Saturdays['TOTAL'][n] = (f[n+1]+f[n+2])/2
        elif f[n] == 0 and a[n+1] == 0 and n == 1:
            BaseLine_Saturdays['TOTAL'][n] = (f[n-1]+f[n+2])/2
        elif f[n] == 0:
            no_ceros = np.count_nonzero(pd.DataFrame({'TOTAL': [f[n-2],f[n-1],f[n+1],f[n+2]]})['TOTAL'].to_numpy())
            suma = sum(pd.DataFrame({'TOTAL': [f[n-2],f[n-1],f[n+1],f[n+2]]})['TOTAL'].to_numpy())
            media_movil = suma/(no_ceros)
            BaseLine_Saturdays['TOTAL'][n] = media_movil      
        elif f[lenF] == 0 and n == lenF:
            BaseLine_Saturdays['TOTAL'][n] = (f[n-1]+f[n-2])/2

    #TRANSFORMACIÓN DE VALORES IGUALES A CERO
    #Días Domingos
    g = BaseLine_Sundays['TOTAL'].to_numpy()
    lenG =len(g)-1
    RangeG =  range(0,lenG)

    for n in RangeG:
        if g[n] == 0 and n == 0:
            BaseLine_Sundays['TOTAL'][n] = (g[n+1]+g[n+2])/2
        elif g[n] == 0 and g[n+1] == 0 and n == 1:
            BaseLine_Sundays['TOTAL'][n] = (g[n-1]+g[n+2])/2
        elif g[n] == 0:
            no_ceros = np.count_nonzero(pd.DataFrame({'TOTAL': [g[n-2],g[n-1],g[n+1],g[n+2]]})['TOTAL'].to_numpy())
            suma = sum(pd.DataFrame({'TOTAL': [g[n-2],g[n-1],g[n+1],g[n+2]]})['TOTAL'].to_numpy())
            media_movil = suma/(no_ceros)
            BaseLine_Sundays['TOTAL'][n] = media_movil      
        elif g[lenG] == 0 and n == lenG:
            BaseLine_Sundays['TOTAL'][n] = (g[n-1]+g[n-2])/2

    #TRANSFORMACIÓN DE VALORES IGUALES A CERO
    #Días Festivos-No domingos
    h = BaseLine_Holydays['TOTAL'].to_numpy()
    lenH =len(h)-1
    RangeH =  range(0,lenH)

    for n in RangeH:
        if h[n] == 0 and n == 0:
            BaseLine_Holydays['TOTAL'][n] = (h[n+1]+h[n+2])/2
        elif h[n] == 0 and h[n+1] == 0 and n == 1:
            BaseLine_Mondays['TOTAL'][n] = (h[n-1]+h[n+2])/2
        elif h[n] == 0:
            no_ceros = np.count_nonzero(pd.DataFrame({'TOTAL': [h[n-2],h[n-1],h[n+1],h[n+2]]})['TOTAL'].to_numpy())
            suma = sum(pd.DataFrame({'TOTAL': [h[n-2],h[n-1],h[n+1],h[n+2]]})['TOTAL'].to_numpy())
            media_movil = suma/(no_ceros)
            BaseLine_Holydays['TOTAL'][n] = media_movil      
        elif h[lenH] == 0 and n == lenH:
            BaseLine_Holydays['TOTAL'][n] = (h[n-1]+h[n-2])/2

    #################################################
    #ELIMINIACIÓN DE MAYOR Y MENOR CONSUMO
    #################################################

    #ELiminación para días laborales
    BaseLine_LaboralDay  = BaseLine_LaboralDay.loc[BaseLine_LaboralDay['TOTAL'] != BaseLine_LaboralDay['TOTAL'].max()]
    BaseLine_LaboralDay  = BaseLine_LaboralDay.loc[BaseLine_LaboralDay['TOTAL'] != BaseLine_LaboralDay['TOTAL'].min()]

    #Eliminación para sábados y domingos (Para festivos no se elimina nada según regulación)
    BaseLine_Saturdays   = BaseLine_Saturdays.loc[BaseLine_Saturdays['TOTAL'] != BaseLine_Saturdays['TOTAL'].max()]
    BaseLine_Saturdays   = BaseLine_Saturdays.loc[BaseLine_Saturdays['TOTAL'] != BaseLine_Saturdays['TOTAL'].min()]
    BaseLine_Sundays     = BaseLine_Sundays.loc[BaseLine_Sundays['TOTAL'] != BaseLine_Sundays['TOTAL'].max()]
    BaseLine_Sundays     = BaseLine_Sundays.loc[BaseLine_Sundays['TOTAL'] != BaseLine_Sundays['TOTAL'].min()]

    #################################################
    # TRANSFORMACIÓN DE VALORES ATÍPICOS
    #################################################

    #Días Lunes
    Q1_BaseLine_LaboralDay  = BaseLine_LaboralDay.quantile(.25)
    Q2_BaseLine_LaboralDays = BaseLine_LaboralDay.quantile(.50)
    Q3_BaseLine_LaboralDay  = BaseLine_LaboralDay.quantile(.75)
    IQR  = Q3_BaseLine_LaboralDay - Q1_BaseLine_LaboralDay
    Xmin = Q1_BaseLine_LaboralDay - 1.5*IQR
    Xmax = Q1_BaseLine_LaboralDay + 1.5*IQR

    a = BaseLine_LaboralDay['TOTAL'].to_numpy()
    lenA   = len(a)
    RangeA = range(0,lenA)

    for n in RangeA:
        if ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == 0):
            BaseLine_LaboralDay['TOTAL'][n] = (a[n+1]+a[n+2])/2
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == 1):
            BaseLine_LaboralDay['TOTAL'][n] = (a[n-1]+a[n+1]+a[n+2])/3    
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == (lenA-2)):
            BaseLine_LaboralDay['TOTAL'][n] = (a[n-1]+a[n-2]+a[n+1])/3
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == (lenA-1)):
            BaseLine_LaboralDay['TOTAL'][n] = (a[n-1]+a[n-2])/2
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])):
            BaseLine_LaboralDay['TOTAL'][n] = (a[n-1]+a[n-2]+a[n+1]+a[n+2])/4

    #Días Sábados
    Q1_BaseLine_Saturdays = BaseLine_Saturdays.quantile(.25)
    Q2_BaseLine_Saturdays = BaseLine_Saturdays.quantile(.50)
    Q3_BaseLine_Saturdays = BaseLine_Saturdays.quantile(.75)
    IQR  = Q3_BaseLine_Saturdays - Q1_BaseLine_Saturdays
    Xmin = Q1_BaseLine_Saturdays - 1.5*IQR
    Xmax = Q1_BaseLine_Saturdays + 1.5*IQR

    a = BaseLine_Saturdays['TOTAL'].to_numpy()
    lenA   = len(a)
    RangeA = range(0,lenA)

    for n in RangeA:
        if ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == 0):
            BaseLine_Saturdays['TOTAL'][n] = (a[n+1]+a[n+2])/2
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == 1):
            BaseLine_Saturdays['TOTAL'][n] = (a[n-1]+a[n+1]+a[n+2])/3    
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == (lenA-2)):
            BaseLine_Saturdays['TOTAL'][n] = (a[n-1]+a[n-2]+a[n+1])/3
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == (lenA-1)):
            BaseLine_Saturdays['TOTAL'][n] = (a[n-1]+a[n-2])/2
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])):
            BaseLine_Saturdays['TOTAL'][n] = (a[n-1]+a[n-2]+a[n+1]+a[n+2])/4


    #Días Domingos
    Q1_BaseLine_Sundays = BaseLine_Sundays.quantile(.25)
    Q2_BaseLine_Sundays = BaseLine_Sundays.quantile(.50)
    Q3_BaseLine_Sundays = BaseLine_Sundays.quantile(.75)
    IQR  = Q3_BaseLine_Sundays - Q1_BaseLine_Sundays
    Xmin = Q1_BaseLine_Sundays - 1.5*IQR
    Xmax = Q1_BaseLine_Sundays + 1.5*IQR

    a = BaseLine_Sundays['TOTAL'].to_numpy()
    lenA   = len(a)
    RangeA = range(0,lenA)

    for n in RangeA:
        if ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == 0):
            BaseLine_Sundays['TOTAL'][n] = (a[n+1]+a[n+2])/2
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == 1):
            BaseLine_Sundays['TOTAL'][n] = (a[n-1]+a[n+1]+a[n+2])/3    
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == (lenA-2)):
            BaseLine_Sundays['TOTAL'][n] = (a[n-1]+a[n-2]+a[n+1])/3
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == (lenA-1)):
            BaseLine_Sundays['TOTAL'][n] = (a[n-1]+a[n-2])/2
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])):
            BaseLine_Sundays['TOTAL'][n] = (a[n-1]+a[n-2]+a[n+1]+a[n+2])/4


    #Días festivos (no domingos)
    Q1_BaseLine_Holydays = BaseLine_Holydays.quantile(.25)
    Q2_BaseLine_Holydays = BaseLine_Holydays.quantile(.50)
    Q3_BaseLine_Holydays = BaseLine_Holydays.quantile(.75)
    IQR  = Q3_BaseLine_Holydays - Q1_BaseLine_Holydays
    Xmin = Q1_BaseLine_Holydays - 1.5*IQR
    Xmax = Q1_BaseLine_Holydays + 1.5*IQR

    a = BaseLine_Holydays['TOTAL'].to_numpy()
    lenA   = len(a)
    RangeA = range(0,lenA)

    for n in RangeA:
        if ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == 0):
            BaseLine_Holydays['TOTAL'][n] = (a[n+1]+a[n+2])/2
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == 1):
            BaseLine_Holydays['TOTAL'][n] = (a[n-1]+a[n+1]+a[n+2])/3    
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == (lenA-2)):
            BaseLine_Holydays['TOTAL'][n] = (a[n-1]+a[n-2]+a[n+1])/3
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])) & (n == (lenA-1)):
            BaseLine_Holydays['TOTAL'][n] = (a[n-1]+a[n-2])/2
        elif ((a[n] < Xmin['TOTAL']) | (a[n] > Xmax['TOTAL'])):
            BaseLine_Holydays['TOTAL'][n] = (a[n-1]+a[n-2]+a[n+1]+a[n+2])/4

    #3  Cálculo del error de estimación de la LBC

    #3.2 Sacar promedios simples por días
    #Calcula la energía promedio  por día de la semana 
    #En la regulación se conoce a cada Curve_Average como la "estimación de la LBC diaria".
    SimpleAverage_summary    = BaseLine_summary[['TOTAL']].mean(axis=0)
    SimpleAverage_LaboralDay = BaseLine_LaboralDay[['TOTAL']].mean(axis=0)
    SimpleAverage_Saturday   = BaseLine_Saturdays[['TOTAL']].mean(axis=0)
    SimpleAverage_Sunday     = BaseLine_Sundays[['TOTAL']].mean(axis=0)
    SimpleAverage_Holiday    = BaseLine_Holydays[['TOTAL']].mean(axis=0)

    #3.3 Número de datos n del día tipo t: laboral, sábado, domingo o festivo,del total de la muestra m.
    m              = len(BaseLine_summary['TOTAL'])
    n_Laboralday   = len(BaseLine_LaboralDay['TOTAL'])
    n_Saturday     = len(BaseLine_Saturdays['TOTAL'])
    n_Sunday       = len(BaseLine_Sundays['TOTAL'])
    n_Holiday      = len(BaseLine_Holydays['TOTAL'])

    #3.4 Cáluclo intermedio para estimar un parámetro de la función RRMSE
    Sumatory_Laboralday = 0
    for n in range(n_Laboralday):
        Sumatory_Laboralday = Sumatory_Laboralday + (SimpleAverage_LaboralDay-list(BaseLine_LaboralDay['TOTAL'])[n])**2

    Sumatory_Saturday = 0
    for n in range(n_Saturday):
        Sumatory_Saturday = Sumatory_Saturday + (SimpleAverage_Saturday-list(BaseLine_Saturdays['TOTAL'])[n])**2

    Sumatory_Sunday = 0
    for n in range(n_Sunday):
        Sumatory_Sunday  = Sumatory_Sunday + (SimpleAverage_Sunday-list(BaseLine_Sundays['TOTAL'])[n])**2

    Sumatory_Holiday = 0
    for n in range(n_Holiday):
        Sumatory_Holiday  = Sumatory_Holiday + (SimpleAverage_Holiday-list(BaseLine_Holydays['TOTAL'])[n])**2    


    #3.4 Cálculo del error de estimación de la LBC (se calculará como el valor relativo de la
    #raíz cuadrada del error medio cuadrático de los datos, RRMSE (por sus siglas en inglés)

    RRMSE_Laboralday   = math.sqrt((1/n_Laboralday)*(Sumatory_Laboralday))/((1/n_Laboralday)*sum(BaseLine_LaboralDay['TOTAL']))
    RRMSE_Saturday     = math.sqrt((1/n_Saturday)*(Sumatory_Saturday))/((1/n_Saturday)*sum(BaseLine_Saturdays['TOTAL']))
    RRMSE_Sunday       = math.sqrt((1/n_Sunday)*(Sumatory_Sunday))/((1/n_Sunday)*sum(BaseLine_Sundays['TOTAL']))
    RRMSE_Holiday      = math.sqrt((1/n_Holiday)*(Sumatory_Holiday))/((1/n_Holiday)*sum(BaseLine_Holydays['TOTAL']))

    #3.5 Cálculo de factor de corrección de la linea base para cada día

    if RRMSE_Laboralday <= 0.05:
        LaboraldayCorrectionFactor = 1
    elif 0.05 < RRMSE_Laboralday <= 0.2:
        LaboraldayCorrectionFactor = (1-RRMSE_Laboralday)
    elif 0.2 < RRMSE_Laboralday:
        LaboraldayCorrectionFactor = 0
    else:
        Print("Hay un error en el cálculo de la RRMSE de los días laborales")

    if RRMSE_Saturday <= 0.05:
        SaturdayCorrectionFactor = 1
    elif 0.05 < RRMSE_Saturday <= 0.2:
        SaturdayCorrectionFactor = (1-RRMSE_Saturday)
    elif 0.2 < RRMSE_Saturday:
        SaturdayCorrectionFactor = 0
    else:
        Print("Hay un error en el cálculo de la RRMSE del Sábado")

    if RRMSE_Sunday <= 0.05:
        SundayCorrectionFactor = 1
    elif 0.05 < RRMSE_Sunday <= 0.2:
        SundayCorrectionFactor = (1-RRMSE_Sunday)
    elif 0.2 < RRMSE_Sunday:
        SundayCorrectionFactor = 0
    else:
        Print("Hay un error en el cálculo de la RRMSE del Domingo")

    if RRMSE_Holiday <= 0.05:
        HolidayCorrectionFactor = 1
    elif 0.05 < RRMSE_Holiday <= 0.2:
        HolidayCorrectionFactor = (1-RRMSE_Holiday)
    elif 0.2 < RRMSE_Holiday:
        HolidayCorrectionFactor = 0
    else:
        Print("Hay un error en el cálculo de la RRMSE de los festivos no domingos")

    #3.5 Cálculo de la LBC diaria

    LBC_Laboralday  = LaboraldayCorrectionFactor*SimpleAverage_LaboralDay
    LBC_Saturday    = SaturdayCorrectionFactor*SimpleAverage_Saturday
    LBC_Sunday      = SundayCorrectionFactor*SimpleAverage_Sunday
    LBC_Holiday     = HolidayCorrectionFactor*SimpleAverage_Holiday

    #3.6 Estimación de la línea base de consumo de energía horaria
    #No entendí
    #XXXXXXXXXXXXXXXXXXXXXXX TRBAJAR AQUI XXXXXXXXXXXXXXXXXXXXXXXXXXXXXX

    #4. VERIFICACIÓN DE DEMANDA DESCONECTADA
    #4.1 verificación de la reducción de consumo de la demanda 
    #desconectada por un usuario se realizará según la siguiente expresión:
    #XXXXXXXXXXXXXXXXXXXXXXX TRBAJAR AQUI XXXXXXXXXXXXXXXXXXXXXXXXXXXXXX


    #5 GUARDAR INFORMACIÓN CALCULADA DE LA FRONTERA EN ARCHIVO EXCEL

    # File directory
    FileDirectory = file
    
    # File directory
    FileDirectory = FileDirectory[:-5]+"_Results_CREG146_version_"+TodayDate+".xlsx"
    print(FileDirectory)

    # Create a Pandas Excel writer using XlsxWriter as the engine.
    writer = pd.ExcelWriter(FileDirectory, engine='xlsxwriter')


    #1. Estimación de la LBC diaria
    Estimacion_LBC_Diaria = pd.concat([SimpleAverage_LaboralDay, 
                                       SimpleAverage_Saturday,
                                       SimpleAverage_Sunday,
                                       SimpleAverage_Holiday], 
                                       keys=['Laboral','Sábado','Domingo','Holidays'],
                                       names =['Día','Estimacion_LBC_Diaria'])

    Estimacion_LBC_Diaria.to_excel(writer, sheet_name='Estimacion_LBC_Diaria')

    #2. Número de datos n del día tipo t
    dias        = ['Laboral','Sábado','Domingo','Holidays']
    Num_dias    = [n_Laboralday,n_Saturday,n_Sunday,n_Holiday]
    Numero_dias = pd.DataFrame()
    Numero_dias['Día'] = dias
    Numero_dias['Número de días'] = Num_dias
    Numero_dias.set_index('Día')
    Numero_dias.to_excel(writer, sheet_name='n_días')

    #3. RRMSE diario
    dias     = ['Laboral','Sábado','Domingo','Holiday']
    RRMSE    = [RRMSE_Laboralday, RRMSE_Saturday,RRMSE_Sunday,RRMSE_Holiday]
    RRMSE_dias = pd.DataFrame()
    RRMSE_dias['Día'] = dias
    RRMSE_dias['RRMSE día'] = RRMSE
    RRMSE_dias.set_index('Día')
    RRMSE_dias.to_excel(writer, sheet_name='RRMSE_dias')

    #4 CorrectionFactors
    dias                 = ['Laboral','Sábado','Domingo','Holiday']
    CorrectionFactors    = [LaboraldayCorrectionFactor, SaturdayCorrectionFactor,SundayCorrectionFactor,HolidayCorrectionFactor]
    CorrectionFactor = pd.DataFrame()
    CorrectionFactor['Día'] = dias
    CorrectionFactor['Factores de corrección'] = CorrectionFactors
    CorrectionFactor.set_index('Día')
    CorrectionFactor.to_excel(writer, sheet_name='Factores de corrección')

    #5 LBC diaria
    dias    = ['Laboral','Sábado','Domingo','Holiday']
    LBCs    = [int(LBC_Laboralday), int(LBC_Saturday), int(LBC_Sunday),int(LBC_Holiday)]
    LBC = pd.DataFrame()
    LBC['Día'] = dias
    LBC['Factores de corrección'] = LBCs
    LBC.set_index('Día')
    LBC.to_excel(writer, sheet_name='LBC diario')


    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
```
